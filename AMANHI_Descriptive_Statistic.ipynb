```python
import pandas as pd

# Set display options for better formatting
pd.set_option('display.expand_frame_repr', False)
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

# Filter only live births
df_alive = df[df["BIRTH_OUTCOME1"] == 1]
df_alive["age_death_b1"].fillna(30, inplace=True)  # Assume survival for missing values
df_alive["NEONATAL_DEATH"] = (df_alive["age_death_b1"] <= 28).astype(int)

# Define the bins and labels for PW_AGE
age_bins = [0, 20, 25, 30, 35, float('inf')]
age_labels = ['Under 20', '20-25', '26-30', '31-35', 'Over 35']
df_alive['PW_AGE_GROUP'] = pd.cut(df_alive['PW_AGE'].astype(float), bins=age_bins, labels=age_labels, right=False)

# Define the bins and labels for education levels
df_alive['PW_EDUCATION'] = df_alive['PW_EDUCATION'].fillna(-1)  # Assign a placeholder for missing values
edu_bins = [-1, 6, 10, 12, float('inf')]
edu_labels = ['Primary_Education', 'Secondary_Education', 'Advanced_Level', 'High_Education']
df_alive['education_category'] = pd.cut(df_alive['PW_EDUCATION'], bins=edu_bins, labels=edu_labels, right=True, include_lowest=True)

# Define the bins and labels for GAGEBIRTH
gage_bins = [0, 259, 294, float('inf')]
gage_labels = ['Preterm', 'Term', 'Post-term']
df_alive['GAGEBRTH_Group'] = pd.cut(df_alive['GAGEBRTH'], bins=gage_bins, labels=gage_labels)

# Define the bins and labels for maternal weight (MAT_WEIGHT)
weight_bins = [0, 50, 70, 90, float('inf')]
weight_labels = ['Underweight', 'Normal', 'Overweight', 'Obese']
df_alive['MAT_WEIGHT_Group'] = pd.cut(df_alive['MAT_WEIGHT'], bins=weight_bins, labels=weight_labels)

# Selected binary feature
binary_feature = 'NEONATAL_DEATH'

# Function to create variation tables with total rows
def create_variation_table(group_col, categories):
    table = pd.DataFrame(index=categories, columns=[f'{binary_feature}_Has', f'{binary_feature}_No'])

    table[f'{binary_feature}_Has'] = df_alive.groupby(group_col, observed=False)[binary_feature].sum()
    table[f'{binary_feature}_No'] = df_alive.groupby(group_col, observed=False)[binary_feature].apply(lambda x: x.size - x.sum())

    table = table.fillna(0).astype(int)
    table['Category'] = table.index
    table['Group_Type'] = group_col
    table = table.reset_index(drop=True)

    # Compute total row for the group
    total_row = table.iloc[:, :-2].sum().to_frame().T  # Sum of all 'Has' and 'No' columns
    total_row['Category'] = 'Total'
    total_row['Group_Type'] = group_col

    return pd.concat([table, total_row], ignore_index=True)  # Append total row

# Creating tables for each group type
variation_table_age_group = create_variation_table('PW_AGE_GROUP', age_labels)
variation_table_education_category = create_variation_table('education_category', edu_labels)
variation_table_gagebrth_group = create_variation_table('GAGEBRTH_Group', gage_labels)
variation_table_weight_group = create_variation_table('MAT_WEIGHT_Group', weight_labels)

# Splitting tables by Group_Type and storing in a dictionary
grouped_tables = {
    'PW_AGE_GROUP': variation_table_age_group,
    'education_category': variation_table_education_category,
    'GAGEBRTH_Group': variation_table_gagebrth_group,
    'MAT_WEIGHT_Group': variation_table_weight_group
}

# Display each table separately
for group, table in grouped_tables.items():
    print(f"\nGroup: {group}")
    print(table.to_string(index=False))
```

```python
import pandas as pd

# Set display options for better formatting
pd.set_option('display.expand_frame_repr', False)
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

# Define the bins and labels for PW_AGE
age_bins = [0, 20, 25, 30, 35, float('inf')]
age_labels = ['Under 20', '20-25', '26-30', '31-35', 'Over 35']
df['PW_AGE_GROUP'] = pd.cut(df['PW_AGE'].astype(float), bins=age_bins, labels=age_labels, right=False)

# Define the bins and labels for education levels
# Fill missing values for education before binning
df['PW_EDUCATION'] = df['PW_EDUCATION'].fillna(-1)  # Assign a placeholder for missing values

# Define education bins and labels
edu_bins = [-1, 6, 10, 12, float('inf')]  # Include -1 to catch missing values
edu_labels = ['Primary_Education', 'Secondary_Education', 'Advanced_Level', 'High_Education']

df['education_category'] = pd.cut(df['PW_EDUCATION'], bins=edu_bins, labels=edu_labels, right=True, include_lowest=True)


# Define the bins and labels for GAGEBIRTH
gage_bins = [0, 259, 294, float('inf')]  # Preterm (<37 weeks), Term (37-42 weeks), Post-term (>42 weeks)
gage_labels = ['Preterm', 'Term', 'Post-term']
df['GAGEBRTH_Group'] = pd.cut(df['GAGEBRTH'], bins=gage_bins, labels=gage_labels)

# Define the bins and labels for maternal weight (MAT_WEIGHT)
weight_bins = [0, 50, 70, 90, float('inf')]  # Underweight (<50 kg), Normal (50-70 kg), Overweight (70-90 kg), Obese (>90 kg)
weight_labels = ['Underweight', 'Normal', 'Overweight', 'Obese']
df['MAT_WEIGHT_Group'] = pd.cut(df['MAT_WEIGHT'], bins=weight_bins, labels=weight_labels)

# Selected binary features
binary_features = ['LABOUR_HTN', 'LABOUR_24', 'APH', 'BIRTH_OUTCOME']

# Function to create variation tables with total rows
def create_variation_table(group_col, categories):
    table = pd.DataFrame(index=categories, columns=[f'{feature}_Has' for feature in binary_features] +
                         [f'{feature}_No' for feature in binary_features])

    for feature in binary_features:
        table[f'{feature}_Has'] = df.groupby(group_col, observed=False)[feature].sum()
        table[f'{feature}_No'] = df.groupby(group_col, observed=False)[feature].apply(lambda x: x.size - x.sum())

    table = table.fillna(0).astype(int)
    table['Category'] = table.index
    table['Group_Type'] = group_col
    table = table.reset_index(drop=True)

    # Compute total row for the group
    total_row = table.iloc[:, :-2].sum().to_frame().T  # Sum of all 'Has' and 'No' columns
    total_row['Category'] = 'Total'
    total_row['Group_Type'] = group_col

    return pd.concat([table, total_row], ignore_index=True)  # Append total row

# Creating tables for each group type
variation_table_age_group = create_variation_table('PW_AGE_GROUP', age_labels)
variation_table_education_category = create_variation_table('education_category', edu_labels)
variation_table_gagebrth_group = create_variation_table('GAGEBRTH_Group', gage_labels)
variation_table_weight_group = create_variation_table('MAT_WEIGHT_Group', weight_labels)

# Splitting tables by Group_Type and storing in a dictionary
grouped_tables = {
    'PW_AGE_GROUP': variation_table_age_group,
    'education_category': variation_table_education_category,
    'GAGEBRTH_Group': variation_table_gagebrth_group,
    'MAT_WEIGHT_Group': variation_table_weight_group
}

# Display each table separately
for group, table in grouped_tables.items():
    print(f"\nGroup: {group}")
    print(table.to_string(index=False))  # Print table without row index
```
